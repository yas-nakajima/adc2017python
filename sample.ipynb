### IRAFを使ってみる (1) 

pythonからIRAFを使うにはpyrafを利用します。  
pyrafを(便利に)使うためには、  

- ホームディレクトリにirafというディレクトリを作成
- そのディレクトリ内で mkirafを実行し、login.cl を作成
- 必要に応じてそのlogin.clを編集  

をしておきます。 こうしておくと、login.clでの設定がpyraf内に反映されます。また、~/iraf/uparm/にパラメータが保存されます。

```python
from pyraf import iraf 
```

これでirafのタスクをpythonで関数として使うことができます。

#### DS9にFITSデータを表示してみる

別のターミナルからds9を立ち上げておく。

```python
iraf.display('./010604.ffiles/hf0001.fits', 1)
```

```python
iraf.display('./010604.ffiles/hf0002.fits', 2)
```

#### imexamも使ってみよう

```python
iraf.imexam()
```

imexamでは、rとかeコマンドを使うとさらに別のグラフィックウィンドウが現れる。

```python
iraf.imexam('./010604.ffiles/hf0001.fits', 1) 
```

#### imstat 

```python
iraf.imstat.unlearn()  #  パラメータをデフォルト値に
```

```python
iraf.imstat('./010604.ffiles/hf0001.fits')
```

#### eparでパラメータ設定 
iraf.epar('タスク名') で、IRAFでおなじみのeparのパラメータ設定の画面が別ウィンドウで開く。GUIでパラメータ設定ができる。同じGUIのウインドウでHELPも読めます。

```python
iraf.epar('imstat')
```

#### 変数としてパラメータ設定

こちらだと、使ったパラメータがこのノートブックに残るので、あとになって「この処理でどんなパラメータ使ったっけ？」となったときに助けになります。

```python
iraf.imstat.fields = 'midpt, mean, stddev'
iraf.imstat.lower = 0
iraf.imstat.upper = 10000
```

```python
iraf.imstat('./010604.ffiles/hf0001.fits')
```

```python
print iraf.imstat.fields
```

```python
iraf.imstat.images = './010604.ffiles/hf0001.fits'
```

```python
iraf.imstat()
```
