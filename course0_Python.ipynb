### インデント

#### forループ 

C言語と比べるとずいぶん勝手がちがいます。  
1. インデントされたブロックが、繰り返される部分です。
2. リストの各要素を順番に繰り返し変数に代入して使います。  

```python
for i in [0,1,2,3]:
    print (i)
```

<img src='./img/for_loop.png', width='500'> 


```python
sum = 0
for i in range(11):  #   range()関数は0から引数の一つ前までの整数を順に返す
    print (i)
    sum += i
print (sum)
```

#### if文  

こちらもインデントによって処理を行うブロックが区別されることを覚えてください。  
<img src='./img/if_else_colon.png', width='500'> 



```python
a = 21
if a > 20:
    print('Large')
else:
    print('small')
```

### モジュールと関数

標準の組み込み関数(print()、range()、open()など)をのぞいて、関数を使用する際には、親となるモジュールをimportする必要があります。

mathモジュールのsqrt()関数を使う場合には次のようにします。  


```python
import math
```

```python
math.sqrt(30)
```
